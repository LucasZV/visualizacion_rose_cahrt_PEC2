```python
import pandas as pd
```

```python
accidentes_23 = pd.read_csv("./data/2023_accidents_gu_bcn.csv")
accidentes_24 = pd.read_csv("./data/2024_accidents_gu_bcn.csv")

accidentes = pd.concat([accidentes_23, accidentes_24])

accidentes_23_causa_conductor = pd.read_csv("./data/2023_accidents_causa_conductor_gu_bcn_.csv")
accidentes_24_causa_conductor = pd.read_csv("./data/2024_accidents_causa_conductor_gu_bcn.csv")
accidentes_23_causa = pd.read_csv("./data/2023_accidents_causes_gu_bcn_.csv")
accidentes_24_causa = pd.read_csv("./data/2024_accidents_causes_gu_bcn.csv")
accidentes_23_tipo = pd.read_csv("./data/2023_accidents_tipus_gu_bcn.csv")
accidentes_24_tipo = pd.read_csv("./data/2024_accidents_tipus_gu_bcn.csv")
accidentes_23_vehiculo = pd.read_csv("./data/2023_accidents_vehicles_gu_bcn.csv")
accidentes_24_vehiculo = pd.read_csv("./data/2024_accidents_vehicles_gu_bcn.csv")
```

```python
# Causa conductor
accidentes_23_causa_conductor = accidentes_23_causa_conductor[["Numero_expedient", "Descripcio_causa_mediata"]]
accidentes_23_causa_conductor = accidentes_23_causa_conductor.rename(columns= {"Descripcio_causa_mediata": "causa_conductor"})

accidentes_24_causa_conductor = accidentes_24_causa_conductor[["Número_expedient", "Descripcio_causa_mediata"]]
accidentes_24_causa_conductor = accidentes_24_causa_conductor.rename(columns= {"Descripcio_causa_mediata": "causa_conductor", "Número_expedient": "Numero_expedient"})

accidentes_conductor = pd.concat([accidentes_23_causa_conductor, accidentes_24_causa_conductor])
```

```python
# Causa
accidentes_23_causa = accidentes_23_causa[["Numero_expedient", "Descripcio_causa_mediata"]]
accidentes_23_causa = accidentes_23_causa.rename(columns= {"Descripcio_causa_mediata": "causa"})

accidentes_24_causa = accidentes_24_causa[["Numero_expedient", "Descripcio_causa_mediata"]]
accidentes_24_causa = accidentes_24_causa.rename(columns= {"Descripcio_causa_mediata": "causa"})

accidentes_causa = pd.concat([accidentes_23_causa, accidentes_24_causa])
```

```python
# Tipo
accidentes_23_tipo = accidentes_23_tipo[["Numero_expedient", "Descripcio_tipus_accident"]]
accidentes_23_tipo = accidentes_23_tipo.rename(columns= {"Descripcio_tipus_accident": "tipo_accidente"})

accidentes_24_tipo = accidentes_24_tipo[["Numero_expedient", "Descripcio_tipus_accident"]]
accidentes_24_tipo = accidentes_24_tipo.rename(columns= {"Descripcio_tipus_accident": "tipo_accidente"})

accidentes_tipo = pd.concat([accidentes_23_tipo, accidentes_24_tipo])
```

```python
# Vehiculo
accidentes_23_vehiculo = accidentes_23_vehiculo[["Numero_expedient", "Descripcio_tipus_vehicle", "Descripcio_model", "Descripcio_marca", "Descripcio_color", "Descripcio_carnet", "Antiguitat_carnet"]]
accidentes_23_vehiculo = accidentes_23_vehiculo.rename(columns= {"Descripcio_causa_vianant": "causa_peaton"})

accidentes_24_vehiculo = accidentes_24_vehiculo[["Numero_expedient", "Descripcio_tipus_vehicle", "Descripcio_model", "Descripcio_marca", "Descripcio_color", "Descripcio_carnet", "Antiguitat_carnet"]]
accidentes_24_vehiculo = accidentes_24_vehiculo.rename(columns= {"Descripcio_causa_vianant": "causa_peaton"})

accidentes_vehiculo = pd.concat([accidentes_23_vehiculo, accidentes_24_vehiculo])
```

```python
# Merge de los datasets
accidentes = pd.merge(accidentes, accidentes_conductor, how = "left", on= "Numero_expedient")
accidentes = pd.merge(accidentes, accidentes_causa, how = "left", on= "Numero_expedient")
accidentes = pd.merge(accidentes, accidentes_tipo, how = "left", on= "Numero_expedient")
accidentes = pd.merge(accidentes, accidentes_vehiculo, how = "left", on= "Numero_expedient")

accidentes = accidentes.fillna(0)
```

```python
def calcular_gravedad(x):
    if x['Numero_morts'] > 0:
        return "Mortal"
    elif x['Numero_lesionats_greus'] > 0:
        return "Grave"
    elif x['Numero_lesionats_lleus'] > 0:
        return "Leve"
    elif x['Numero_victimes'] == 0:
        return "Sin víctimas"
    else:
        return "Sin víctimas"

accidentes['gravedad'] = accidentes.apply(calcular_gravedad, axis=1)

accidentes["fecha"] = accidentes["NK_Any"].astype(str) + "-" + accidentes["Mes_any"].astype(str) + "-" + accidentes["Dia_mes"].astype(str)
accidentes["fecha"] = pd.to_datetime(accidentes["fecha"])
```

```python
accidentes.to_csv("./data/accidentes_full.csv", index=False)
```
